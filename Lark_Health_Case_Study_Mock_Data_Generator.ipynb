### height

```python
# Add height
min_height_female, max_height_female, mean_female, std_female = 56, 72, 65, 3
min_height_male, max_height_male, mean_male, std_male = 60, 78, 70, 3

#generate distributions by sex
dist_female = stats.truncnorm((min_height_female - mean_female) / std_female, (max_height_female - mean_female) / std_female, loc = mean_female, scale = std_female)
dist_female_list = dist_female.rvs(10000)
dist_male = stats.truncnorm((min_height_male - mean_male) / std_male, (max_height_male - mean_male) / std_male, loc = mean_male, scale = std_male)
dist_male_list = dist_male.rvs(10000)

def height_gen(sex):
  """Pass in sex as 'female' or 'male'. Return sex specific height from distribution common to US population. Values in inches."""
  if sex == 'female':
    gen_height = random.choice(dist_female_list)
  else:
    gen_height = random.choice(dist_male_list)
  gen_height = int(round(gen_height))
  return gen_height

users['height'] = users['sex'].map(height_gen)
users.head()
```

### start_date

```python
def random_skewed(minimum, maximum, skewness):
  """Accept min, max, and skewness. Return list of values in skewed distribution. """
  num_values = 10000
  random = stats.skewnorm.rvs(a = skewness,loc = maximum, size = num_values)  #Skewnorm function
  random = random - min(random)  # Shift the set so the minimum value is equal to zero.
  random = random / max(random)  # Standadize all the values between 0 and 1. 
  random = random * maximum  # Multiply the standardized values by the maximum value.
  random = random.round().astype(int)
  mask = np.logical_and(random >= minimum, random <= maximum)
  random = random[mask]
  return random
```

```python
# Check works for month
m = random_skewed(1,12,-4)
print(np.median(m), np.unique(m, return_counts=True))
sns.histplot(m)
```

```python
# Check works for date
d = random_skewed(1,28,8)
print(np.median(d), np.unique(d, return_counts=True))
sns.histplot(d)
```

```python
def gen_month_int():
  """Output random number between 1 and 12 with left skew (most months are towards end of year)"""
  month_gen = random.choice(random_skewed(1,12,-4))
  return month_gen

def get_day_int():
  """Output random number between 1 and 28 with righ skew (most days are at beginning of month"""
  date_gen = random.choice(random_skewed(1,28,8))
  return date_gen

def get_year_int():
  """Output date from list with weights"""
  dates = [2020,2021,2022]
  weights = [0.2, 0.5, 0.3]
  year_gen = np.random.choice(dates, p=weights)
  return year_gen
  
def get_start_date():
  """Output full date as string using gen_month_int(), get_day_int(), and get_year_int()"""
  start_date_gen = f'{gen_month_int()}/{get_day_int()}/{get_year_int()}'
  return start_date_gen
```

```python
users['start_date'] = [get_start_date() for user in range(num_users)]
```

### start_BMI
- Weight BMI categories are: 18.5-24.9 (normal), 25-29.9 (overweight), 30-34.9 (obese I), 35-39.9 (obese II), and >=40 (obese III)

```python
def get_start_bmi(bmi_category):
  """Output start bmi from list with weights"""

  if bmi_category == 2:
    start_bmi_gen = round(random.uniform(25, 29.9),1)
  elif bmi_category == 3:
    start_bmi_gen = round(random.uniform(30, 34.9),1)
  elif bmi_category == 4:
    start_bmi_gen = round(random.uniform(35, 39.9),1)
  elif bmi_category == 5:
    start_bmi_gen = round(random.uniform(40, 60),1)
  else:
    start_bmi_gen = 0

  return start_bmi_gen

users['start_BMI'] = users['start_BMI_category'].map(get_start_bmi)
users.head()
```

### end_weight
* let's assume that women lose more weight than the men here in a normal distribution between -2% and 12% bodyweight with average of 9 and std of 2
* let's assume men lose less than the wormen here in a normal distribution between -5% and 12% bodyweight with average of 5 and std of 3 (**cohortA**)
* let's assume that after some new feature is applied, men close to how much wormen lose here in a normal distribution between -2% and 15% bodyweight with average of 7 and std of 2 (**cohortB**)
* new_weight = old_weight * (1 - minloss|or|maxloss)


```python
# Add height
min_weightloss_f, max_weightloss_f, mean_weightloss_f, std_weightloss_f = -2, 12, 9, 2  # Female
min_weightloss_m, max_weightloss_m, mean_weightloss_m, std_weightloss_m = -5, 12, 5, 3  # Male CohortA
#min_weightloss_m, max_weightloss_m, mean_weightloss_m, std_weightloss_m = -2, 15, 7, 2  # Male CohortB

#generate distributions by sex
dist_weightloss_f = stats.truncnorm((min_weightloss_f - mean_weightloss_f) / std_weightloss_f, (max_weightloss_f - mean_weightloss_f) / std_weightloss_f, loc = mean_weightloss_f, scale = std_weightloss_f)
dist_weightloss_f_list = dist_weightloss_f.rvs(10000)
dist_weightloss_m = stats.truncnorm((min_weightloss_m - mean_weightloss_m) / std_weightloss_m, (max_weightloss_m - mean_weightloss_m) / std_weightloss_m, loc = mean_weightloss_m, scale = std_weightloss_m)
dist_weightloss_m_list = dist_weightloss_m.rvs(10000)

def get_end_weight(sex, weight):
  """Accept sex and start_weight as parameter, output end_weight based on distribution of losses predefined above."""
  
  if sex == 'female':
    weightloss_percentage_f = random.choice(dist_weightloss_f_list)
    end_weight_gen = int(round(weight * (1 - (weightloss_percentage_f/100))))
  else:
    weightloss_percentage_m = random.choice(dist_weightloss_m_list)
    end_weight_gen = int(round(weight * (1 - (weightloss_percentage_m/100))))

  return end_weight_gen
```

```python
users['end_weight'] = users.apply(lambda row: get_end_weight(row['sex'],row['start_weight']),axis=1)
users.head()
```

### start_a1c

**A1c Test %**
* Diabetes: 6.5% 
* Prediabetes: 5.7 – 6.4%	
* Normal: Below 5.7% 

```python
def get_start_a1c(bmi_category):
  """Accept a bmi_category from 1 = normal to 5 = obeseIII. Return an a1c value in an appropriate range."""
  
  if bmi_category == 1:  # Normal
    start_a1c_gen = round(random.uniform(0, 5.7),1)
  elif bmi_category == 2:  # Overweight
    start_a1c_gen = round(random.uniform(5.7, 6.4),1)
  elif bmi_category == 3:  # ObeseI
    start_a1c_gen = round(random.uniform(6, 8),1)
  elif bmi_category == 4:  # ObeseII
    start_a1c_gen = round(random.uniform(8, 10),1)
  elif bmi_category == 5:  # ObeseIII
    start_a1c_gen = round(random.uniform(10, 14),1) 
  else:
    start_a1c_gen = 0

  return start_a1c_gen
```

```python
users['start_a1c'] = users['start_BMI_category'].map(get_start_a1c)
users.head()
```

### goal_a1c
"After adjustment for antidiabetic medication intake, for every 10% weight loss, the predicted reduction in A1C% was 0.81. Conclusions: Intentional weight loss of 10% can potentially decrease A1C% by 0.81 among patients with type 2 DM."

So let's assume if our target is 5% weight loss then the a1c target decrease is 0.4 to 0.6

```python
def get_goal_a1c(start_a1c):
  """Subtract a randomly generated number between 0.4 and 0.6 from start_a1c"""
  goal_a1c_gen = round(start_a1c - round(random.uniform(0.4, 0.6),1),1)
  return goal_a1c_gen
```

```python
users['goal_a1c'] = users['start_a1c'].map(get_goal_a1c)
users.head()
```

### end_a1c

```python
def get_end_a1c(start_weight, end_weight, start_a1c):
  """Accept a bmi_category from 1 = normal to 5 = obeseIII. Return an a1c value in an appropriate range.
  For every 1% bodyweight lost, reduce a1c by 0.08."""
  
  weight_dropped = start_weight - end_weight
  percent_weightlost = 100 * weight_dropped / start_weight
  end_a1c_gen = round(start_a1c - (percent_weightlost * 0.08),1)

  return end_a1c_gen
```

```python
users['end_a1c'] = users.apply(lambda row: get_end_a1c(row['start_weight'], row['end_weight'], row['start_a1c']), axis=1)
users.head()
```

### total_conversations
* Let's create a scenerio where individuals that hit their goal weight ( end_weight <= goal_weight ) had more conversations, meals_logged, weight_measurements, missions_completed, mental_healthchecks, percentage_healthy_meals_consumed, and net_promoter_score.

```python
# Define distributions for those that hit their target weight by the end
min_hit, max_hit, mean_hit, std_hit = 300, 450, 390, 20

# Define the distributions for those that failed to hit their target weight by the end
min_failed, max_failed, mean_failed, std_failed = 30, 150, 50, 15

# Generate the distributions from the definitions above
dist_hit = stats.truncnorm((min_hit - mean_hit) / std_hit, (max_hit - mean_hit) / std_hit, loc = mean_hit, scale = std_hit)
dist_hit_list = dist_hit.rvs(10000)
dist_failed = stats.truncnorm((min_failed - mean_failed) / std_failed, (max_failed - mean_failed) / std_failed, loc = mean_failed, scale = std_failed)
dist_failed_list = dist_failed.rvs(10000)
```

```python
def get_total_conversations(end_weight, goal_weight):
  """Accept end_weight and goal_weight. Return total conversations"""
  hit_goal = end_weight <= goal_weight  # True/False
  if hit_goal == True:
    total_conversations_gen = round(random.choice(dist_hit_list))
  else:
    total_conversations_gen = round(random.choice(dist_failed_list))
  return total_conversations_gen
```

```python
# Re-Define hit and failed

# Define distributions for those that hit their target weight by the end
min_hit, max_hit, mean_hit, std_hit = 40, 100, 60, 15

# Define the distributions for those that failed to hit their target weight by the end
min_failed, max_failed, mean_failed, std_failed = 0, 39, 20, 8

# Generate the distributions from the definitions above
dist_hit = stats.truncnorm((min_hit - mean_hit) / std_hit, (max_hit - mean_hit) / std_hit, loc = mean_hit, scale = std_hit)
dist_hit_list = dist_hit.rvs(10000)
dist_failed = stats.truncnorm((min_failed - mean_failed) / std_failed, (max_failed - mean_failed) / std_failed, loc = mean_failed, scale = std_failed)
dist_failed_list = dist_failed.rvs(10000)
```

```python
def get_percentage_healthy_meals_consumed(end_weight, goal_weight):
  """Accept end_weight and goal_weight. Return percentage_healthy_meals_consumed"""
  hit_goal = end_weight <= goal_weight  # True/False
  if hit_goal == True:
    percentage_healthy_meals_consumed_gen = round(random.choice(dist_hit_list))
  else:
    percentage_healthy_meals_consumed_gen = round(random.choice(dist_failed_list))
  return percentage_healthy_meals_consumed_gen
```

```python
users['percentage_healthy_meals_consumed'] = users.apply(lambda row: get_percentage_healthy_meals_consumed(row['end_weight'], row['goal_weight']), axis=1)
users.head()
```

### net_promoter_score

* "How likely is it that you would recommend Lark to a friend or colleague?" [0-10 scale]

Respondents are grouped as follows:
* Promoters (score 9-10) are loyal enthusiasts who will keep buying and refer others, fueling growth.
* Passives (score 7-8) are satisfied but unenthusiastic customers who are vulnerable to competitive offerings.
* Detractors (score 0-6) are unhappy customers who can damage your brand and impede growth through negative word-of-mouth.
* Subtracting the percentage of Detractors from the percentage of Promoters yields the Net Promoter Score, which can range from a low of -100 (if every customer is a Detractor) to a high of 100 (if every customer is a Promoter)

```python
def get_nps(end_weight, goal_weight):
  """Accepts end_weight and goal_weight, if goal is hit then promoter else detractor or passive"""
  hit_goal = end_weight <= goal_weight  # True/False
  if hit_goal == True:
    nps_gen = round(random.uniform(9, 10)) #promoter
  else:
    nps_gen = round(random.uniform(0, 8)) #detractor or passive
  return nps_gen
```

```python
users['net_promoter_score'] = users.apply(lambda row: get_nps(row['end_weight'],row['goal_weight']),axis=1)
users.head()
```
